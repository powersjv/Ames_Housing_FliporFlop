# Visualizations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/train_featengin.csv')

#df.drop(columns=['Unnamed: 0'], inplace=True)

#allows me to scroll (Credit to Andy)
pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
```

```python
ols_features = ['overall_qual_gr_liv_area', 
                'ms_zoning_RL', 'ms_zoning_RH', 
                'fireplaces', 
                'kitchen_qual', 
                'functional', 
                'baths', 
                'ms_zoning_FV', 
                'condition_rank',
                'ms_zoning_RM',
                'exter_cond', 
                'overall_qual',
                'bsmt_num', 
                'exter_num', 
                'neighborhood_rank', 
                'totrms_abvgrd', 
                'garage_space', 
                'lot_area', 
                'yr_remod_or_built', 
                'deck_or_porch', 
                'ms_zoning_I (all)', 
                'ms_zoning_C (all)']
```

```python
sns.pairplot(df, x_vars = ols_features, y_vars = ['saleprice'])
```

```python
sns.scatterplot(x = df['overall_qual'], y = df['gr_liv_area']);
```

```python
plt.figure(figsize = (13, 10))
sns.boxplot(x = df['overall_qual'], y = df['gr_liv_area'], palette = 'summer')
plt.title('Overall Quality and Above Ground Living Area')
plt.xlabel('Overall Quality  (1 - Poor, 10 - Excellent)')
plt.ylabel('Above Ground Living Area(sqft)')
plt.rc('axes', titlesize=18)
plt.rc('axes', labelsize=14)
plt.rc('ytick', labelsize=14)
plt.ylim(0,4000)
plt.savefig("../images/ols_features_boxplot_overallqual_vs_abvgrlivarea.jpg", bbox_inches="tight", dpi=100);
```

```python
df_kitchen = df.groupby(['kitchen_qual']).mean()
df_kitchen
```

```python
plt.figure(figsize = (5, 5))
sns.barplot(x = df_kitchen.index, y = 'saleprice', data=df_kitchen, color = '#f9bf3e')
plt.ylim(0, 400000)
plt.ylabel('Sale Price ($)')
plt.title('Kitchen Quality vs Sale Price')
plt.xlabel('Kitchen Quality (1 - Poor, 5 - Excellent)')
plt.rc('axes', titlesize=15)
plt.rc('axes', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.savefig("../images/ols_features_kitchen_vs_saleprice.jpg", bbox_inches="tight", dpi=100);
```

```python
df_baths = df.groupby(['baths']).mean()
```

```python
plt.figure(figsize = (5, 5))
sns.barplot(x = df_baths.index, y = 'saleprice', data=df_baths, color = '#9ad7d2')
plt.ylim(0, 400000)
plt.ylabel('Sale Price ($)')
plt.title('Bathrooms vs Sale Price')
plt.xlabel('Number of Baths    (Full + Half(0.5)')
plt.rc('axes', titlesize=15)
plt.rc('axes', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.savefig("../images/ols_features_baths_vs_saleprice.jpg", bbox_inches="tight", dpi=100);
```

```python
df_fireplaces = df.groupby(['fireplaces']).mean()
```

```python
plt.figure(figsize = (5, 5))
sns.barplot(x = df_fireplaces.index, y = 'saleprice', data=df_fireplaces, color = '#9ad7d2')
plt.ylim(0, 400000)
plt.ylabel('Sale Price ($)')
plt.title('Fireplaces vs Sale Price')
plt.xlabel('Number of Fireplaces')
plt.rc('axes', titlesize=15)
plt.rc('axes', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.savefig("../images/ols_features_fireplaces_vs_saleprice.jpg", bbox_inches="tight", dpi=100);
```
